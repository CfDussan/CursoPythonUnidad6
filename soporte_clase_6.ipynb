### Componentes claves en Matplotlib

- Figura: Es el contenedor principal de la visualización. Piensa en la figura como una hoja en blanco en el que se dibujarán todos los elementos de nuestro gráfico. Una figura en Matplotlib puede contener uno o más ejes, y cada elemento gráfico se coloca en relación con la figura completa.

- Ejes: Los ejes representan un área de trazado individual dentro de la figura. Un gráfico de Matplotlib puede contener múltiples ejes, que permiten la presentación de diferentes conjuntos de datos con escalas propias. Los ejes incluyen dos (o tres en gráficos 3D) líneas de eje (denominadas habitualmente X e Y) que contienen las marcas (o ticks) y etiquetas que definen la escala de los datos representados.

- Es importante comprender las caracteristicas de ambos así como sus diferencias dado que en ello radica la posibilidad de sacarle mayor provecho a la herramienta que es altamente customizable si sabemos donde buscar y/o editar cada elemento

### Histogramas, Diagramas de Dispersión y Gráficos de Barras.

- Los histogramas son útiles para visualizar la distribución de un conjunto de datos numéricos a través de barras, donde cada barra representa la frecuencia (número de ocurrencias) de los datos en un intervalo específico. Básicamente, permiten ver la frecuencia con la que ocurren diferentes rangos de valores dentro de un conjunto. Es particularmente útil para explorar la forma de la distribución subyacente de los datos, como identificar si la distribución es normal, sesgada, bimodal, etc.

- Un gráfico o diagrama de dispersión es una representación visual que muestra la relación entre dos conjuntos de datos o variables numéricas. Cada punto en el gráfico refleja los valores de dos variables, una en el eje X y otra en el eje Y. Este tipo de gráfico es fundamental para la exploración de datos, ya que permite identificar visualmente correlaciones, tendencias y agrupaciones, así como detectar comportamientos atípicos entre las variables.

- Los gráficos de barras son útiles para comparaciones categóricas, permitiéndonos ver cómo varían los valores a través de diferentes grupos o categorías. Es ideal para visualizar distribuciones de datos o comparaciones entre categorías donde la orientación vertical de las barras facilita la lectura de los valores representados y permite una fácil comparación de alturas de barra.

```python
# Histograma

plt.figure(
    figsize=(10,6),
    facecolor="#FDED26", # Color de relleno de la figura
    edgecolor="white" # Color del borde de la figura
)

plt.hist(
    data=df,
    x="Salario USD",
    bins=30,
    range=(df["Salario USD"].min() * 1.1, df["Salario USD"].max() * 0.9), # Permite excluir valores atípicos u "Outliers"
    # density=True # Permite ver las repeticiones en densidad de probabilidad en lugar de absolutos
    # cumulative=True # Hist acumulativo en lugar de un Hist simple
    color="gray"
)

plt.title(
    label="Distribución Salarial en USD",
    loc="center",
    pad=10,
    fontdict=title_styles # type: ignore
)

plt.show()
```

### Más sobre personalización de los gráficos

```python
# Calculando el salario medio por país
mean_salaries = df.groupby('Ubicación')['Salario USD'].mean()

# Estableciendo un estilo
plt.style.use('classic') # Otras opciones: ggplot, seaborn-v0_8, classic, bmh, dark_background, fivethirtyeight

# Creando la figura y los ejes
fig, ax = plt.subplots()

# Personalizando y creando el diagrama de barras
ax.bar(mean_salaries.index, mean_salaries.values, color='skyblue', label='Salario Medio')

# Añadiendo leyendas y anotaciones
ax.legend(loc='upper right') # Añade la leyenda en el margen superior derecho
ax.annotate('Punto destacado', xy=('Argentina', 5400), xytext=('Argentina', 5700),
             arrowprops=dict(facecolor='black', shrink=0.05)) # Inserto una anotación en mi gráfico

# Estableciendo etiquetas y títulos de ejes
plt.xlabel('País')
plt.ylabel('Salario Medio (USD)')
ax.set_ylabel('Salario Medio')

# Añadiendo el título al gráfico
ax.set_title('Salario medio por país')

# Añadiendo una cuadrícula
plt.grid(True)

# Personalizando las marcas del eje x
plt.xticks(rotation=45)

# Mostrando el gráfico
plt.show()
```
