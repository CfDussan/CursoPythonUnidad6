#### Creación de la figura y los ejes: 

plt.subplots() es una función que crea una nueva figura y un conjunto de ejes en esta figura.

fig es una referencia a la figura completa. Puedes pensar en la figura como en el contenedor nivel más alto donde se dibujan todos los gráficos.

ax es una referencia a los ejes de la figura creada. En este caso, se está creando un único conjunto de ejes, pero subplots puede configurarse para crear múltiples gráficos en una sola figura.

#### Creación el histograma propiamente dicho:

ax.hist() como ya vimos es la función que crea el histograma. Se llama sobre el objeto de ejes ax que se creó en el paso anterior.

df['Salario USD'] especifica los datos que se van a utilizar para el histograma, en este caso, es la columna 'Salario USD' del DataFrame df. Esto significa que el histograma mostrará la distribución de los valores de salario en USD contenidos en df.

bins=10 indica que los datos deben ser divididos en 10 contenedores (bins) de igual ancho. Los bins son esenciales en un histograma, ya que determinan cómo se agrupan los datos en diferentes rangos. La elección del número de bins puede afectar significativamente la interpretación de la distribución de los datos.

## Diagrama de dispersión

### ax.scatter(). 

```python
ax.scatter(x, y, s=None, c=None, marker=None, cmap=None)
# s = tamaño
# c = color
# marker = la forma en que mis datos se va a visualizar
# cmap = si quiero tener un color diferenciado para mis puntos
```

Un gráfico o diagrama de dispersión es una representación visual que muestra la relación entre dos conjuntos de datos o variables numéricas

Para crear un diagrama de dispersión en Python, y más específicamente con Matplotlib, utilizamos la función ax.scatter(). 

ax.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None, **kwargs)

Por ejemplo, si queremos analizar la relación entre los salarios en USD y las horas de capacitación, podríamos hacerlo de la siguiente manera:

```python
# Creando la figura y los ejes
fig, ax = plt.subplots()

# Creando el diagrama de dispersión
ax.scatter(df['Salario USD'], df['Horas de Capacitación Anual'])

# Mostrando el gráfico
plt.show()
```

El diagrama de dispersión se crea mediante el método ax.scatter(), y se deben definir los dos campos que se van a graficar y para los cuales se desea mostrar su relación.

```python
# Creando la figura y los ejes
fig, ax = plt.subplots()

# Creando el diagrama de dispersión
ax.scatter(df['Salario USD'], df['Horas de Capacitación Anual'], c = 'y', marker = '^')
# Cambiamos color y el tipo de objeto que se muestra, en este caso un trinagulo.

# Mostrando el gráfico
plt.show()
```

## Gráfico de barras

### ax.bar()

```python
x.bar(x, height, width=0.8, bottom=None, **kwargs)
# x las posiciones de las categorias en la barra x
# height la altura que va a tener cada una de las barras
# width ancho de cada una
# bottom posicion de inicio, en este caso en el eje y
# **kwargs
```

Los gráficos de barras son útiles para comparaciones categóricas, permitiéndonos ver cómo varían los valores a través de diferentes grupos o categorías.

ax.bar() - Gráficos de barras verticales
Sintaxis:
ax.bar(x, height, width=0.8, bottom=None, **kwargs)

ax.bar() es ideal para visualizar distribuciones de datos o comparaciones entre categorías donde la orientación vertical de las barras facilita la lectura de los valores representados y permite una fácil comparación de alturas de barra.

### ax.barh() - Gráficos de barras horizontales

```python
ax.barh(y, width, height=0.8, left=None, **kwargs)
# width ancho de cada una
# height la altura que va a tener cada una de las barras
```

Sintaxis:
ax.barh(y, width, height=0.8, left=None, **kwargs)

ax.barh() se utiliza cuando es más conveniente o claro presentar los datos horizontalmente, especialmente útil para etiquetas largas de categoría o cuando se quiere enfatizar la duración o el ancho en lugar de la altura.

Veamos un ejemplo en base al caso descrito en la clase: supongamos que queremos comparar los salarios promedio de los distintos países:

```python
# Con lo aprendido en clases anteriores, obtenemos el salario medio por país
mean_salaries = df.groupby('Ubicación')['Salario USD'].mean()

# Creando la figura y los ejes
fig, ax = plt.subplots()

# Creando el gráfico de barras (en este caso verticales)
ax.bar(mean_salaries.index, mean_salaries.values)

# Mostrando el gráfico
plt.show()
```

```python
# Con lo aprendido en clases anteriores, obtenemos el salario medio por país
mean_salaries = df.groupby('Ubicación')['Salario USD'].mean()

# Creando la figura y los ejes
fig, ax = plt.subplots()

# Creando el gráfico de barras (en este caso verticales)
ax.barh(mean_salaries.index, mean_salaries.values)

# Mostrando el gráfico
plt.show()
```
