```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Importar el dataset
df = pd.read_csv('dataset_techtechlam.csv')
```

```python
# Revisar las primeras filas para verificar que se cargo correctamente
print(df.head())
```

```python
# Supongamos que queremos ver la evolución del salario promedio anual a lo largo de los años.
# Agrupar los datos por año y calcular el salario promedio
#salario_promedio_anual = df.groupby('Año')['Salario USD'].mean()
salario_promedio_anual = df.groupby('Año')['Salario USD'].mean()

# Imprimimos el salario promedio
print(np.around(salario_promedio_anual, decimals=2))
```

```python
# Crear un gráfico simple usando .plot()
plt.figure(figsize=(10, 6))
salario_promedio_anual.plot(kind='line', title='Evolución del Salario Promedio Anual en IT en América Latina')
# kind='line' especificamos un grafico de lineas
# figsize=(10, 6) define el tamaño de la figura del gráfico (ancho x alto en pulgadas). 
# title asigna un título al gráfico.

# Mostrar el gráfico
plt.show()
# plt.show() se utiliza para mostrar el gráfico generado.

```

```python
# Aprendiendo
salario_promedio_rol = df.groupby('Rol')['Salario USD'].mean()
print(np.around(salario_promedio_rol,decimals=2))
```

```python
salario_promedio_nivel = df.groupby('Nivel')['Salario USD'].mean()
print(np.around(salario_promedio_nivel,decimals=2))
```

```python
# Aprendiendo mas
salario_promedio_rol_año = df.groupby(['Año','Rol'])['Salario USD'].mean().reset_index()
salario_promedio_nivel_año = df.groupby(['Año','Nivel'])['Salario USD'].mean().reset_index()
```

```python
print(salario_promedio_nivel_año)
```

```python
# Jugando
plt.figure(figsize=(10, 6))
salario_promedio_anual.plot(kind='line', title='Evolución del Salario Promedio Anual en IT en América Latina')
salario_promedio_rol_año.plot(kind='bar')
salario_promedio_nivel_año.plot(kind='bar')
plt.show()
```

```python

# Crear un gráfico de líneas
plt.figure(figsize=(12, 8))
for rol in salario_promedio_rol_año['Rol'].unique():
    subset = salario_promedio_rol_año[salario_promedio_rol_año['Rol'] == rol]
    plt.plot(subset['Año'], subset['Salario USD'], marker='o', label=rol)

plt.title('Salario Promedio por Año y Rol')
plt.xlabel('Año')
plt.ylabel('Salario USD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Crear un gráfico de líneas
plt.figure(figsize=(12, 8))
for nivel in salario_promedio_nivel_año['Nivel'].unique():
    subset = salario_promedio_nivel_año[salario_promedio_nivel_año['Nivel'] == nivel]
    plt.plot(subset['Año'], subset['Salario USD'], marker='o', label=nivel)

plt.title('Salario Promedio por Año y Nivel')
plt.xlabel('Año')
plt.ylabel('Salario USD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Crear un gráfico de barras
plt.figure(figsize=(12, 8))
for nivel in salario_promedio_nivel_año['Nivel'].unique():
    subset = salario_promedio_nivel_año[salario_promedio_nivel_año['Nivel'] == nivel]
    plt.bar(subset['Año'], subset['Salario USD'], label=nivel)

plt.title('Salario Promedio por Año y Nivel')
plt.xlabel('Año')
plt.ylabel('Salario USD')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Agrupar por 'Rol' y 'Año' y calcular la media del salario
salario_promedio_rol_ano = df.groupby(['Rol', 'Año'])['Salario USD'].mean().reset_index()

# Redondear los valores a 2 decimales
salario_promedio_rol_ano['Salario USD'] = np.around(salario_promedio_rol_ano['Salario USD'], decimals=2)

# Crear un gráfico de barras
fig, ax = plt.subplots(figsize=(12, 8))

# Generar una barra por cada combinación de Rol y Año
roles = salario_promedio_rol_ano['Rol'].unique()
for i, rol in enumerate(roles):
    subset = salario_promedio_rol_ano[salario_promedio_rol_ano['Rol'] == rol]
    ax.bar(subset['Año'] + i*0.1, subset['Salario USD'], width=0.1, label=rol)  # desplazamiento de barras

ax.set_title('Salario Promedio por Año y Rol')
ax.set_xlabel('Año')
ax.set_ylabel('Salario USD')
ax.legend()
ax.grid(True)
plt.xticks(salario_promedio_rol_ano['Año'].unique())
plt.show()
```
