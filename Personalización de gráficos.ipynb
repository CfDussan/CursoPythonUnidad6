# Personalización de gráficos

La personalización de tus gráficos es una tarea muy importante para lograr que la información sea clara y comprensible, además de dar un aspecto profesional 

La función .plot() es extremadamente versátil y puede tomar una variedad de atributos para personalizar la presentación de los gráficos.

Vamos a utilizar un gráfico de barras para ilustrar los conceptos, utilizaremos un código integral partiendo del tema anterior, y luego desarrollaremos los métodos utilizados de manera detallada.

```python
# Calculando el salario medio por país
mean_salaries = df.groupby('Ubicación')['Salario USD'].mean()

# Estableciendo un estilo
plt.style.use('ggplot')

# Creando la figura y los ejes
fig, ax = plt.subplots()

# Personalizando y creando el diagrama de barras
ax.bar(mean_salaries.index, mean_salaries.values, color='skyblue', label='Salario Medio')

# Añadiendo leyendas y anotaciones
ax.legend(loc='upper right')
ax.annotate('Punto destacado', xy=('Argentina', 65000), xytext=('Argentina', 70000),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Estableciendo etiquetas y títulos de ejes
plt.xlabel('País')
plt.ylabel('Salario Medio (USD)')
ax.set_ylabel('Salario Medio')

# Añadiendo el título al gráfico
ax.set_title('Salario medio por país')

# Añadiendo una cuadrícula
plt.grid(True)

# Personalizando las marcas del eje x
plt.xticks(rotation=45)

# Mostrando el gráfico
plt.show()
```

Pasos que realiza el script:

1
Calcula el salario medio por país utilizando un grupo por operación en el DataFrame df.

2
Establece un estilo visual para el gráfico usando plt.style.use().

3
Crea un gráfico de barras vertical (ax.bar()) que representa el salario medio por país.

4
Añade una leyenda y una anotación para destacar un punto específico en el gráfico.

5
Establece las etiquetas para los ejes X e Y y añade un título al gráfico con ax.set_title('Salario medio por país').

6
Personaliza las marcas del eje X para mejorar la legibilidad y añade una cuadrícula para facilitar la interpretación de los valores.

### Desarrollo de métodos, funciones y atributos

#### 1️⃣ Calculando el salario medio por país

Se utiliza groupby para agrupar los datos por la columna 'Ubicación', y luego se calcula el salario medio utilizando .mean() para la columna 'Salario USD'. El resultado es una Serie de pandas (mean_salaries) donde el índice son los países (Ubicación) y los valores son los salarios medios correspondientes. 

#### 2️⃣ Configuración de estilo

plt.style.use('ggplot')

Se selecciona el estilo 'ggplot' para el gráfico, que es uno de los estilos predefinidos en Matplotlib. Este estilo proporciona una estética visualmente agradable con una cuadrícula de fondo gris y colores pastel.

Sintaxis básica:

plt.style.use(nombre_del_estilo)

#### 3️⃣ Creación de la figura y los ejes

fig, ax = plt.subplots()

Se crea una figura y un conjunto de ejes usando plt.subplots(). 

Sintaxis básica:

fig, ax = plt.subplots()
Esta línea de código, en su forma más simple, crea una figura y un único conjunto de ejes.

Sin embargo, plt.subplots() es altamente configurable y permite especificar el número de filas y columnas de subgráficos, su disposición, tamaño de la figura, entre otros aspectos, por ejemplo:

fig, axs = plt.subplots(nrows=2, ncols=2)

En este ejemplo, se crearán 4 subgráficos (2x2) dentro de una única figura, y axs será un arreglo de objetos Axes que se pueden manipular de forma individual.

#### 4️⃣ Calculando el salario medio por país

ax.bar(mean_salaries.index, mean_salaries.values, color='skyblue', label='Salario Medio')

Tal como vimos en el tema anterior, se utiliza ax.bar() para crear un gráfico de barras verticales. Los índices de mean_salaries (los países) se usan como las etiquetas en el eje X, y los valores (los salarios medios) como las alturas de las barras. 

En este caso, se personaliza el color de las barras a azul cielo ('skyblue') y se añade una etiqueta 'Salario Medio' para la leyenda.

#### 5️⃣ Calculando el salario medio por país

Leyendas y anotaciones:
ax.legend(loc='upper right')
ax.annotate('Punto destacado', xy=('Argentina', 65000), xytext=('Argentina', 70000),
             arrowprops=dict(facecolor='black', shrink=0.05))

Leyenda: ax.legend(loc='upper right') añade una leyenda en la esquina superior derecha. La leyenda ayuda a identificar qué representan las barras.

Conozcamos más sobre la función ax.legend():

ax.legend() es una función en Matplotlib que se utiliza para agregar una leyenda a un gráfico.

Anotación: ax.annotate() se usa para añadir una anotación en el gráfico, señalando un punto de interés. En este caso, señala el salario medio de Argentina con un texto 'Punto destacado' y una flecha.

Sintaxis básica:

ax.annotate(text, xy, xytext=None, xycoords='data', textcoords='data', arrowprops=None, **kwargs)

Etiquetas y títulos:

plt.xlabel('País')
plt.ylabel('Salario Medio (USD)')
ax.set_ylabel('Salario Medio')
ax.set_title('Salario medio por país')

plt.xlabel() es una función en Matplotlib diseñada para establecer el título o etiqueta del eje X en un gráfico. 

plt.ylabel() es una función en Matplotlib que se utiliza para establecer el título o etiqueta del eje Y en un gráfico.

ax.set_ylabel('Valores cuadrados', fontsize=14, color='red')

En este caso, la etiqueta del eje Y se establece con una fuente de tamaño 14 y color rojo, lo que puede ayudar a que el gráfico sea más atractivo o más fácil de leer dependiendo del contexto en el que se presente.

ax.set_ylabel() versus plt.ylabel() 

ax.set_ylabel() y plt.ylabel() son dos funciones que se utilizan para establecer la etiqueta del eje Y en un gráfico, pero difieren en cómo se aplican dentro del contexto de la creación de figuras y ejes en Matplotlib. La diferencia principal entre estas dos funciones radica en su enfoque de orientación hacia el objeto (Object-Oriented) versus el enfoque basado en el estado (state-based o pyplot style).

#### 6️⃣ Ajustes finales

Cuadrícula y rotación de etiquetas:
plt.grid(True)
plt.xticks(rotation=45)

plt.grid(True) añade una cuadrícula para facilitar la lectura de los valores.
plt.xticks(rotation=45) rota las etiquetas del eje X para mejorar la legibilidad.

#### 7️⃣ Visualización

plt.show()

Mostrar el gráfico: finalmente, plt.show() se llama para mostrar el gráfico generado.
